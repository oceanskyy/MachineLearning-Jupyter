## Multinomial Logistic Regression

Multinomial Logistics Regression is used when the prediction of the outcome is a nominal variable with two or more categories that do not have a specific order. This also extends the concept of logistic regression in order to solve multiple possible outcomes with one or more independent variables.

The group has chosen Mutinomial Logistics Regression as one of the machine learning techniques/models since we have decided to use the dataset of Traffic Predictions and to classify them by how severe the traffic is for a day. We do not need to rank the severity of the traffic since the group's aim is to only classify.

multinomial logisctic regression.png

```python
# import scikit-learn's Logistic Regression algorithm
from sklearn.linear_model import LogisticRegression

# Import for evaluating the performance of regression models in machine learning
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
```

*   Normalize the dataset
  - This preprocessing step ensures that all input features are on a similar scale, which aids in the model's convergence speed during training. Multinomial logistic regression computes coefficients (weights) for each feature to make predictions for multiple classes. When features have different scales, larger magnitude features might disproportionately influence the model, leading to biased or less accurate coefficients. Normalizing features to a common scale ensures fair treatment among all features, resulting in more reliable coefficients and better model performance.

    - What 'StandarScaler' does is it standardizes features by scaling them to have a mean of 0 and a standard deviation of 1.
    - Formula:
      where xi is the value, mu is the mean, and sigma is the standard deviation


standardization.png

- fit_transform computes the mean and standard deviation of each feature in the training set 'X_train' and then scales the features accordingly. It then uses the same values acquired to transform or scale the test set 'X_test'.

```python
from sklearn.preprocessing import StandardScaler

# Normalize/Scale the features using StandardScaler
scaler = StandardScaler()
X_trainLR = scaler.fit_transform(X_train)
X_testLR = scaler.transform(X_test)
```

To use multinomial logisitic regression with the LogisticRegression class, enable this functionality by configuring the settings. Specify the "multi_class" parameter as multinomial and select an appropriate solver that supports multinomial logistic regression, such as "sag". (Stochastic Average Gradient)
  - "sag" solver is chosen because it is designed for large-scale problems and could provide good performances while being memory-efficient.

Subsequently, the model is trained using the training data represented by X_trainLR (the scaled or normalized feature data) and y_train (the corresponding target labels). During training, the model learns the relationships between the features and the target classes in order to make predictions.
  - Logits or scores are produced by linear model.
  linear_model.png
  - For Multinomial logistic regression, the **softmax function** is used to compute the probabilities for each class. The softmax function takes the logits as input and produces a probability distribution, which sums up to 1.
  softmax.png
    - e is the base of the natural logarithm (Euler's number)
    - zi is the score or logit for class 'i'
    - K is the total number of classes
    - The denominator represents the sum of exponential scores for all classes


The Cross-Entropy Loss Function is utilized during the training of the Multinomial Logistic Regression model to measure the dissimilarity between the predicted probability distribution and the actual distribution of classes in the training dataset. As the model learns, the goal is to minimize this loss function by adjusting the model's parameters.

During the training phase, the model employs the Gradient Descent algorithm to iteratively update its parameters based on the gradient of the loss function with respect to these parameters. The gradient represents the direction and magnitude of the steepest ascent of the loss function in the parameter space. By computing the gradients and moving in the opposite direction (negative gradient), the model optimizes its parameters to minimize the cross-entropy loss.

However, once the model is trained and the parameters are optimized based on the training data, during the prediction phase, it leverages the learned patterns and relationships to make predictions on new or unseen data. The 'predict' method is then used to generate predicted class labels (y_pred) for the test dataset (X_testLR) based on the knowledge acquired from the training phase.

```python
# Create a multinomial logistic regression model
model = LogisticRegression(multi_class="multinomial", solver="sag")

# Fit the model to the training data
model.fit(X_trainLR, y_train)

# Make predictions on the test set
y_pred = model.predict(X_testLR)
```

```python
# Accessing coefficients (weights) of each feature learned by the model
coefficients = model.coef_[0]
feature_names = X.columns

# Printing coefficients and corresponding feature names
for feature, coef in zip(feature_names, coefficients):
    print(f"Feature: {feature}, Coefficient: {coef}")
```

The coefficients indicate the direction and strength of the influence that each feature has on the target variable in the logistic regression model. Positive coefficients suggest a positive impact on the probability of the target class, while negative coefficients suggest a negative impact.

For instance, CarCount has a positive coefficient, which means as the CarCount increases, the log-odds (or probability) of the predicted target class also increases.

```python
# Calculate the confusion matrix
confusion_matrix_LR = confusion_matrix(y_test, y_pred)
confusion_matrix_LR
```

```python
# visualize the confusion matrix

plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix_LR, annot=True, cmap='Blues', fmt='g',
            xticklabels=['Low', 'Normal', 'Heavy', 'High'],
            yticklabels=['Low', 'Normal', 'Heavy', 'High'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

- Confusion Matrix Values:
  - The diagonal elements (from top to bottom) indicate the number of correctly classified instances for each class.
    - The value at row 1, column 1 (191) indicates that 191 instances were correctly predicted by the model as "Low".

    - High True Positives: 479 instances were correctly predicted as "High"
  - Weakness of the Model

- Compute for the accuracy. It measures the overall correctness of predictions made by the model.

Accuracy = Total of Correct Prediction / Total Prediction

```python
# Calculate accurancy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy for Multinomial Logistic Regression: {accuracy}")
```

 The model accurately predicted the classifications with a 89% correctness rate.

- Compute for the precision. It focuses on the accuracy of positive predictions made by the model.

Precision = True Positive / True Positive + False Positive

```python
# Calculate precision
precision = precision_score(y_test, y_pred, average="weighted")
print(f"Precision for Multinomial Logistic Regression: {precision}")
```

- Compute for the recall. It measures the model's ability to identify all positive instances correctly.

```python
# Calculate recall
recall = recall_score(y_test, y_pred, average="weighted")
print(f"Recall for Multinomial Logistic Regression: {recall}")
```

The model achieved a recall of 89.14%, signifying that it successfully captured 89.14% of all actual positive instances in the dataset.

```python
# Calculate F1 Score
f1score = f1_score(y_test, y_pred, average="weighted")
print(f"F1 score for Multinomial Logistic Regression: {f1score}")
```

The F1 score for the Multinomial Logistic Regression model is 89%, indicating a strong balance between precision and recall in its predictions.

## Random Forest Classifier

The Random Forest Classifier is a supervised machine learning model that can be used for classification tasks. It uses an ensemble learning method that combines different predictions and multiple models to create more accurate predictions. The algorithm builds a forest or ensemble of decisions trees. A prediction is made by searching for the best features among a random subset of features.

The group chose Random Forest as the second technique because of its versatility as it can be used in both classification and regretion tasks and because of the accuracy of the prediction of the algorithm.

```python
# import scikit-learn's Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

# Import for evaluating the performance of regression models in machine learning
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
```

As for the **dataset**, normalizing the data is not needed for Random Forest Classifier because decision trees make splits based on individual feature, not their scale or magnitude. Additionally, in Random Forest, each tree operates independently on random subsets of features and data samples, making combined model less sensitive to variations in feature scales. However, it's essential to maintain parity between both models in terms of preprocessing steps for fair comparison.

```python
from sklearn.preprocessing import StandardScaler

# Normalize/Scale the features using StandardScaler
scaler = StandardScaler()
X_trainRF = scaler.fit_transform(X_train)
X_testRF = scaler.transform(X_test)
```

**Random Forest Classifier model** is created using the RandomForestClassifier class from scikit-learn. The model is configured with 100 decistion trees (n_estimators=10), and 'random_state' parameter is set to 42, ensuring reproducibility by fixing the random number generation seed. These numbers are the numbers commonly used in machine learning.

1. Random Forest is built upon individual decision trees. Each tree is trained on a random subset of a training data and random subset of features.
2. At each split in a tree, a random subset of features is considered for determining the best split. This randomness prevents individual trees from relying too much on a particular feature, enhancing diversity.
3. Each tree in the Random Forest is grown by repeatedly splitting nodes based on the best feature and split point. It aims to maximize information gain or decrease in impurity (Gini impurity or entropy). The growing stops when it reaches a specified depth.
4.  In classification, the mode (most frequent/majority voting) prediction among trees is taken as the overall prediction.


**Model Training**
  - The criterion used for splitting nodes in the decision trees within the Random Forest is '**gini**' (this is the default when not explicitly stated). This refers to the Gini impurity criterion. Gini Impurity serves as a metric utilized in constructing Decision Trees, helps in determining how dataset features should be divided at nodes to form the tree structure.

image.png

Image from learndatascience.com

  - When training a decision tree, the attribute that provides the smallest Gini Impurity is chosen to split the node.

  - The Random Forest model is trained using the training dataset, consisting of feature data (X_train) and their corresponding target labels (y_train). During this stage, the model constructs multiple decision trees by utilizing subsets of the training data, aiming to learn relationships between features and target classes.

Random Forest uses ensemble learning method by combining predictions from multiple individual decision trees. Each decision tree is within the ensemble is trained independetly on random subsets of data, contributing to the overall model's prediction.

```python
# Create a Random Forest Classifier model
modelRF = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
modelRF.fit(X_trainRF, y_train)

# After training, we can make prediction from the model
y_pred_RF = modelRF.predict(X_testRF)
```

```python
# Calculate confusion matrix
confusion_matrix_RF = confusion_matrix(y_test, y_pred_RF)
confusion_matrix_RF
```

```python
# visualize the confusion matrix

plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix_RF, annot=True, cmap='Reds', fmt='g',
            xticklabels=['Low', 'Normal', 'Heavy', 'High'],
            yticklabels=['Low', 'Normal', 'Heavy', 'High'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

# Comparative Analysis


## How the two model works
  - Multinomial Logistic Regression
  - Random Forest Classifier

Logistic Regression, by default, handles solely two-class classification. There are methods to expand its applicability to multi-class problems, such as techniques like **one-vs-rest**, but they require transforming the task into multiple binary classification problems. In contrast, **Multinomial Logistic Regression** is an algorithm that extends the logistic regression model. It involves utilizing softmax function instead of sigmoid function and incorporates cross-entropy loss, thus providing built-in support for multi-class classification.



process_MLR.jpg

Image from dataaspirant.com

- The inputs (X) are the features of the dataset. In our case, its date, time, day of the week, count of each vehicles and total count. These features must be in its correct datatype, which in our case, we converted them earlier to integers.

- Multinomial logistic regression assumes a linear relationship between the input features and the logits, or the output of the linear model,  of the multiple classes. It uses a linear combination of features weighted by coefficients for each class. In machine learning models such as logistic regression or linear models, the concept of "features weighted by coefficients" refers to how the model learns the importance of each feature by assigning coefficients (weights) to them during the training process. These weights are then multiplied to X (inputs in numerical form).

- For the probability estimation, the model applies the softmax function to the ouput of the linear combination. Softmax converts these raw scores into probabilities, ensuring that they sum up to 1 across all classes. The class with the highest probability is predicted.

- The last stage is the Cross-Entropy Loss. During training, it minimizes the cross-entroy loss, which measures the dissimilarity between predicted probabilities and the actual class labels. It serves as a method to measure distance, using the probabilities derived from the softmax function and one-hot encoded matrix. Smaller distance values indicate a closer match to the correct target class, while larger values signify a mismatch with the intended target class.

Next model is the **Random Forest**. It is a well-known machine learning technique used in both classification and regression tasks. It operates through ensemble learning, a method that combines several classifiers to tackle intricate problems and boost tbe model's effectiveness. It is composed of multiple decision trees working on different parts of a dataset. By averaging the results of these trees, it aims to improve the accuracy of predictions and determines the final output based on the most common prediction among them.

process_RFC.png

Image from researchgate.net

Steps in Random Forest Classifier
- First is it randomly select subsets of the training dataset to create multiple datasets for each tree in the forest. This process is known as **bootstrapping**.
- Followed by Growing Decision Trees by randomly selecting a subset of features at each node split to endure diversity among the trees. The splitting of the nodes is based on the chosen criterion, in our case, we used **Gini impurity** to minimize the misclassification.
- After that, voting will be performed. For classification tasks, each tree predicts the class of an input sample. The final prediction is determined by taking a majority vote among all the trees' predictions.

Advantages
  - The strength of Random Forest lies in **combining multiple decision trees**, which makes the prediction more accurate than a single tree.
  - It **reduces the overfitting**. Overfitting means the model learns the dataset too much, to the extent that it stores the random flunctuations rather than learning the patterns and understanding the relationships of the features.
  - It can also provide **feature importance**, which is the measure of feature importance based on how much each feature contributes to reducing impurity or error across the trees.
